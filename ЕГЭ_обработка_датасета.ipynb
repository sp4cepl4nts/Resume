```python
# Дата старта курса: изменим тип данных
df['Дата старта курса'] = pd.to_datetime(df['Дата старта курса'])
df.info()
```

```python
# \r в конце названий некоторых городов
df['Город ученика'].unique()
```

```python
# Отрезаем \r справа от названий некоторых городов
df['Город ученика'] = df['Город ученика'].str.rstrip('\r')
df['Город ученика'].unique()
```

```python
# Имя вместо фамилии, английские буквы
df['Фамилия ученика'].value_counts()
```

```python
# Значение inf, нужно отфильтровать
df['Сколько полных месяцев курса открыто у ученика'].unique()
```

```python
import numpy as np
df = df[~(df['Сколько полных месяцев курса открыто у ученика'] == np.inf)]
df['Сколько полных месяцев курса открыто у ученика'].unique()
```

```python
def determine_wave(row):
    delta = (row['Дата открытия курса ученику'] - row['Дата старта курса']).days
    if delta <= 0:
        return 0
    elif delta <= 7:
        return 1
    elif delta <= 14:
        return 2
    elif delta <= 21:
        return 3
    elif delta <= 28:
        return 4
    else:
        return 5

# Применение функции для создания колонки wave
df['Волна'] = df.apply(determine_wave, axis=1)
```
